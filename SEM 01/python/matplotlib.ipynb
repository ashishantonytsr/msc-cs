# Plots

## `plt.plot()` parameters
`plt.plot(x, y, *args, **kwargs`\
shorthand method: **'color, marker, linestyle'**\


|marker | . |color | . | linestyle | . |
|--- | --- |--- | --- | --- | --- |
|**'o'** |  Circle marker | **'r'** | Red | **'-'** |	Solid line |
|**'v'** |  Downward-pointing triangle marker | **'g'** | Green | **':'** | Dotted line |
|**'^'** |  Upward-pointing triangle marker | **'b'** | Blue |  **'--'** | Dashed line |
|**'<'** |  Left-pointing triangle marker | **'y'** | Yellow |  **'-.'** | Dashed dotted line |
|**'>'** |  Right-pointing triangle marker | **'m'** | Magenta |
|**'s'** |  Square marker | **'c'** | Cyan |  
|**'p'** |  Pentagon marker | **'k'** | Black |
|**'*'** |  Star marker | **'w'** | White |
|**'h'** |  Hexagon1 marker |
|**'H'** |  Hexagon2 marker |
|**'+'** |  Plus marker |
|**'x'** |  X marker |
|**'D'** |  Diamond marker |


```python
x = np.arange(1, 20)
y = np.array([a for a in x])
y1 = np.array([a**2 for a in x])
y2 = np.array([a**(5/2) for a in x])
y3 = np.array([1/a for a in x])

# plt.xlim(-200, 500)
plt.ylim(-10, 500)

plt.plot(x, y, 'r')
plt.plot(x, y1, 'g+')
plt.plot(x, y2, 'bx:')
plt.plot(x, y3, 'yo', ms=1)

plt.show()
```

| `plt.plot()` params | . |
| --- | --- |
| **marker** | Type of marker |
| **ms** |  Marker size |
| **mfc** | Marker facing color |
| **mec** | Marker edge color |
| **ls** |  Linestyle |
| **lw** |  Linewidth |
| **c** |  Linecolor |

```python
x1 = np.arange(1, 10)
y1 = np.arange(5, 50, 5)

plt.plot(x1, y1, ms=5, mec='g', mfc='r', ls='dotted', marker = 'o', lw=2)

plt.show()
```

## pyplot functions
* `plt.title(<str>, loc='right')`
* `plt.fontdict(<dict>)`
* `plt.xlabel(<str>)`
* `plt.ylabel(<str>)`
* `plt.grid(axis=<str>, c=<str>, lw=<float>, ls=<str>)`


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 20)
y = np.array([a**(1.5) for a in x])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.grid(c='r', lw=0.5, ls='-.')

plt.show()
```

## Display multiple plots
`plt.subplot()` is used for plotting multiple plots\
Takes 3 arguments\
1. number of rows (should be same for a set of plots)
2. number of cols (should be same for a set of plots)
3. index of plot, w.r.t rows and cols

```python
x1 = np.arange(23, 100)
y1 = np.array([a**2 for a in x1])
plt.subplot(2, 2, 1)
plt.title('first')
plt.plot(x1, y1, c='r')

x2 = np.arange(100, 18, -2)
y2 = np.array([a**2 for a in x2])
plt.subplot(2, 2, 2)
plt.title('second')
plt.plot(x2, y2, c='g')

x3 = np.arange(1, 25)
y3 = np.array([a**3 for a in x3])
plt.subplot(2, 2, 3)
plt.title('third')
plt.plot(x3, y3, c='b')

plt.show()
```

# Scatter Plots


```python
x1 = np.random.randint(-50, +50, (20,))
y1 = np.random.randint(-50, +50, (20,))
plt.scatter(x1, y1, color='green')

x2 = np.random.randint(-15, +10, (20,))
y2 = np.random.randint(-10, +15, (20,))
plt.scatter(x2, y2, color='red')

plt.show()
```

## plt.scatter() args

|.|.|
|---|---|
|**c**: color| \<str> |
|**s**: size| <int \| float>|
|**alpha**| \<float>|

```python
sizes = np.random.randint(100, 800, (14, ))
colors = np.array(["green","magenta","blue","yellow","red","black","orange"]*2)

x1 = np.random.randint(-50, +50, (14,))
y1 = np.random.randint(-50, +50, (14,))
plt.scatter(x1, y1, c=colors, s=sizes, alpha=0.5)


```

# Bar plots


```python
x1 = np.array(['A', 'B', 'C', 'D', 'E'])
y1 = np.random.randint(20, 100, (5,))

plt.bar(x1, y1, width=.8)
plt.show()
```

# Horizonatal bar plot

```python
x1 = np.array(['a', 'b', 'c', 'd', 'e'])
y1 = np.random.randint(20, 100, (5,))

plt.barh(x1, y1, color='cyan', height=0.7)
plt.show()
```

# Histogram

```python
x1 = np.random.normal(150, 20, 500)

plt.hist(x1)
plt.show()
```

# Pie chart

```python
x1 = np.random.randint(5, 10, 6)

labels = np.array([chr(a) for a in np.random.randint(65, 92, 6)])
explode = np.array([0.1, 0, 0, 0, 0, 0])
colors = np.array(["green","magenta","blue","brown","red","orange"])

plt.pie(x1, labels=labels, explode=explode, colors=colors)
plt.show()
```
